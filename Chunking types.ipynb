```python
#from langchain_ai21 import AI21SemanticTextSplitter

from langchain.chains import ConversationalRetrievalChain
from langchain.chat_models import ChatOpenAI
from langchain_openai import ChatOpenAI
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma, DocArrayInMemorySearch
from langchain_text_splitters import TokenTextSplitter
from langchain.text_splitter import CharacterTextSplitter
```

```python
#Load the PDF document
loader = PyPDFLoader("/Users/pradhikshasuresh/Documents/Python/Space.pdf")
documents = loader.load()
# Initialize the LLM model
llm_model = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0, openai_api_key='My OPEN_API_KEY')

# Splitting using RecursiveCharacterSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
docs = text_splitter.split_documents(documents)
# Create embeddings
embeddings = OpenAIEmbeddings(openai_api_key='My OPEN_API_KEY')
# Create a vector store with Chroma
db = Chroma.from_documents(docs, embeddings)
# Create a retriever
retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 5})
qa = ConversationalRetrievalChain.from_llm(
        llm=llm_model,  
        retriever=retriever, 
    )
```

```python
# Splitting using TikToken
text_splitter1 = TokenTextSplitter(chunk_size=1000, chunk_overlap=150)
docs1 = text_splitter1.split_documents(documents)
# Create embeddings
embeddings = OpenAIEmbeddings(openai_api_key='My OPEN_API_KEY')
# Create a vector store with Chroma
db = Chroma.from_documents(docs1, embeddings)
# Create a retriever
retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 5})
qa1 = ConversationalRetrievalChain.from_llm(
    llm=llm_model,
    retriever=retriever,
)

#splitting using characterslpitting
text_splitter2 = CharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
docs2 = text_splitter2.split_documents(documents)
# Create embeddings
embeddings = OpenAIEmbeddings(openai_api_key='My OPEN_API_KEY')
# Create a vector store with Chroma
db = Chroma.from_documents(docs2, embeddings)
# Create a retriever
retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 5})
qa2 = ConversationalRetrievalChain.from_llm(
    llm=llm_model,
    retriever=retriever,
)

chat_history = []

question="What is light year"

result = qa({"question": question, "chat_history": chat_history})
result1 = qa1({"question": question, "chat_history": chat_history})
result2 = qa2({"question": question, "chat_history": chat_history})

print("Result from RecursiveCharacterSplitter:")
print(result)
print("Result from TikToken:")
print(result1)
print("Result from characterslpitting:")
print(result2)

text="LangChain is a robust framework for building applications powered by large language models."

r_splitter = RecursiveCharacterTextSplitter(chunk_size=10, chunk_overlap=0)
r_splitter.split_text(text)

t_splitter = TokenTextSplitter(chunk_size=, chunk_overlap=0)
t_splitter.split_text(text)

c_splitter = CharacterTextSplitter(chunk_size=10, chunk_overlap=0,separator = ' ')
c_splitter.split_text(text)
```
